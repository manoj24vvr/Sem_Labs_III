### 2. Generate a boxplot to find the distribution of unemployment rate for every industry .

```python
fig, ax = plt.subplots(figsize = (15,10), sharex = True)
ax = df.boxplot(fontsize=10)
ax.set_xlabel("Unemployment rate")
ax.set_title("Distribution of unemployment rate for every industry")
plt.show()
```

### 4. Plot the monthly and yearly trends .

```python
df_by_month = df.resample('M').mean()
df_by_year = df.resample('Y').mean()
start, end = '2001-01', '2010-02'


fig, ax = plt.subplots(figsize=(15,10))
ax.plot(df_by_year.loc[start:end, 'Agriculture'], marker ='.',linestyle='-',linewidth=0.5,color ='black', label = 'Daily')
ax.plot(df_by_month.loc[start:end, 'Agriculture'],marker='o',markersize = 0.8, linestyle='-',color = 'coral', label = 'Monthly')
ax.set_ylabel("Open")
ax.legend()
```

### 5. Apply time series decomposition to your dataset to visualize the trend and seasonality .


```python
import statsmodels.api as sm
plt.rcParams['figure.figsize'] = 11,9
decomposition = sm.tsa.seasonal_decompose(df_by_month['Agriculture'], model = 'Additive')
fig = decomposition.plot()
plt.show()
```
