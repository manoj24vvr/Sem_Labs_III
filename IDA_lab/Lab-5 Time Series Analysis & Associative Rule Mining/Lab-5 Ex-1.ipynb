### 7. Visualize the seasonality of multiple time series and the correlation between each time series in the dataset.

```python
df_month = df.resample("M").mean()
fig, ax = plt.subplots(figsize=(14,6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Agriculture"], width=25, align='center',label = 'Agriculture')
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Education and Health"], width=25, align='center',label = 'Education and Health')
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Finance"], width=25, align='center',label = 'Finance')
plt.xlabel("Month")
plt.ylabel("Unemployment")
plt.legend()
```
